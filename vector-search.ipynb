```python
from langchain_community.document_loaders import TextLoader # Convert raw text into a format langchain can work with
from langchain_text_splitters import CharacterTextSplitter # split the whole doc containing descriptions into meaningful chunks
from langchain_openai import OpenAIEmbeddings # converting those chunks into document embeddings
from langchain_chroma import Chroma # store those in a vector database
```

```python
from dotenv import load_dotenv

load_dotenv(override=True)
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")
```

```python
books.head()
```

```python
books["tagged_description"].to_csv("tagged_description.txt",
                                   sep = "\n",
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(documents, embedding=OpenAIEmbeddings())
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"]==int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(query, top_k=10):
    recs = db_books.similarity_search(query, k=50)
    books_list = []
    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]
    return books[books["isbn13"].isin(books_list)].head(top_k)
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
